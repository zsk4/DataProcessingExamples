```python
# Zachary Katz
# zachary_katz@mines.edu
# 02 February 2024

"""
Load feature csv [Obtained from ] and time to next slip csv.
Filter events that have a time to next slip.
"""

# Imports and Extensions
import pandas as pd

# Load data
feature_df = pd.read_csv("./_Data/features.csv")
time_to_slip_df = pd.read_csv("./_Data/filtered_time_to_next_event.csv", index_col=0)
assert feature_df.shape[0] == time_to_slip_df.shape[0], "Dataframe missing events"

# Merge dataframes and drop events without a time to next slip
merged_df = feature_df.merge(time_to_slip_df, left_index=True, right_index=True)
merged_df = merged_df[merged_df["time_to_next_ev_hr"] != -1]
merged_df.to_csv("./_Data/filtered_features.csv")
merged_df.head()


```
